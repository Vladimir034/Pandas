#### Штаты где безработица выше 50% и мужчин более 50%

#### Процент состава мужчин в населении штата

```python
usa['%men'] = (usa['Men'] / (usa['Men'] + usa['Women']))*100
```

```python
usa[(usa['Unemployment'] > 50) & (usa['%men'] > 50)].head()
```

### Групируем по штатам где безрабоца выше 50 % с мужским населением выше 50%

```python
usa[(usa['Unemployment']>50)&(usa['%men']>50)].groupby('State').head(5).sort_values('TotalPop',ascending=False)
```
